```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,6]
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,"b-.")
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,"r",label="X^3")
plt.plot(x,y2,color="black",label="X^2")
plt.ylabel("Function of X")
plt.xlabel("X")
plt.legend()
plt.text(4,64,"(4,64)")
plt.title("Matplotlib Demo")
plt.show()
```

```python
#pie graph
sizes=[3,4,6,2]
labels=["A","B","C","D"]
colors=["blue","green","red","pink"]
explode=[0.1,0.2,0,0]
plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct="%.1f%%",counterclock=False,startangle= 100)
plt.title("Splitted Data")
plt.show()
```
